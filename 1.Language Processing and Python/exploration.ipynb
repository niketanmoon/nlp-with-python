## Vocabulary

- vocabulary is a set of texts. Since its a set, all the duplicates are dropped

```python
len(text3)
```

```python
sorted(set(text3))
```

> all the uppercase words comes before lowercase

> Each of these characters and words are called as tokens

```python
# getting the vocabulary
len(set(text3))
```

- These 2789 distinct words are also called as word types. A word type is a word considered as a unique item of vocab.

```python
len(text3) / len(set(text3))
```

```python
# counting the frequency of a word
print("Frequency of word smote: ", text3.count("smote"))
# what percentage of the text is taken up by a specific word
print("Percentage of frequency of word: ", 100 * text3.count('smote') / len(text3))
```

```python
print("Frequency of word lol: ", text5.count("lol"))
print("Percentage of frequency of word lol: ", 100 * text5.count('lol') / len(text5))
```

```python
def lexical_diversity(text):
    return len(text) / len(set(text))

def percentage(count, total):
    return 100 * count / total
```

```python
lexical_diversity(text3)
```

```python
percentage(text3.count('smote'), len(text3))
```

## Frequency Distribution

```python
# find top 50 words in the Moby Dick text
fdist1 = FreqDist(text1)
vocabulary1 = fdist1.most_common(50)
vocabulary1
```

```python
## plotting the frequency distribution
fdist1.plot(50, cumulative=True);
```

```python
# words that occur only once
fdist1.hapaxes()
```
